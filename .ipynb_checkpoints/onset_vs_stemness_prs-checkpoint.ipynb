```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr
```

```python
tcga_info = pd.read_csv('../data/paper/clinical_PANCAN_patient_with_followup.tsv', 
                        sep='\t', header=0, index_col=1, dtype=object)
```

```python
prs = pd.read_csv('~/DATA1/Hyoeun/spliceQTL/Analysis/04_PRS_Diff/runPRSice/INPUT/whole_sample/PRAD/whole-GWAS.SUM.all.score.tab', 
                  sep='\t', header=0, index_col=0)
prs.columns = list(prs.columns[1:]) + ['None']
prs = prs.drop('None', axis=1)
```

```python
# mRNAsi
stem = pd.read_csv('../data/TCGA-TARGET-GTEx/mRNA_StemScore.extended.tsv', sep='\t', header=0, index_col=0)
```

```python
stem = pd.read_excel('../data/paper/TCGA_stemness.xlsx', sheet_name=0, header=0, index_col=0)
stem['patient_id'] = stem.index.str.rsplit('-', 4).str[0]
```

```python
stem = stem[stem['cohort'] == 'TCGA']
stem['patient_id'] = stem.index.str.rsplit('-', 1).str[0]
```

# PRS ~ onset correlation

```python
idx = np.intersect1d(tcga_info.index, prs.index)
onset = pd.to_numeric(tcga_info.loc[idx, 'age_at_initial_pathologic_diagnosis'])
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axes = axes.flatten()
disease_list = ['PRAD', 'BRCA', 'LIHC', 'UCEC', 'THCA', 'LUAD', 'COAD', 'ESCA', 'STAD']
for i in range(len(disease_list)):
    disease = disease_list[i]
    prs = pd.read_csv('/home/omics/DATA1/Hyoeun/spliceQTL/Analysis/04_PRS_Diff/runPRSice/INPUT/whole_sample/Tumor/{0}/whole-GWAS.SUM.all.score.tab'.format(disease), 
                      sep='\t', header=0, index_col=0)
    prs.columns = list(prs.columns[1:]) + ['None']
    prs = prs.drop('None', axis=1)
    idx = np.intersect1d(tcga_info.index, prs.index)
    onset = pd.to_numeric(tcga_info.loc[idx, 'age_at_initial_pathologic_diagnosis'])
    r, p = pearsonr(prs.loc[idx, '0.050000'], onset)
    sns.regplot(prs.loc[idx, '0.050000'], onset, ax=axes[i], 
                label='Pearson \nr={0:.2f} p={1:.3}'.format(r, p))
    axes[i].set_xlabel('PRS 0.05')
    axes[i].set_title(disease)
    axes[i].legend()
sns.despine()
```

```python
fig.savefig('../data/TCGA-TARGET-GTEx/figures/20180921/PRAD_BRCA.PRS_0.05_onset_age.png')
fig.savefig('../data/TCGA-TARGET-GTEx/figures/20180921/PRAD_BRCA.PRS_0.05_onset_age.pdf')
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axes = axes.flatten()
disease_list = ['PRAD', 'BRCA', 'LIHC', 'UCEC', 'THCA', 'LUAD', 'COAD', 'ESCA', 'STAD']
status= 11
si = 'mDNAsi'
# Drop [Not Available]
onset = tcga_info[tcga_info['age_at_initial_pathologic_diagnosis'] != '[Not Available]']['age_at_initial_pathologic_diagnosis']
onset = pd.to_numeric(onset)
for i in range(len(disease_list)):
    disease = disease_list[i]
    tmp_stem = stem[(stem['cancer.type'] == disease) & (stem['sample.type'] == status)].set_index('patient_id', drop=False)
    tmp_stem = tmp_stem.drop_duplicates('patient_id')
    idx = np.intersect1d(onset.index, tmp_stem.index)
    r, p = pearsonr(tmp_stem.loc[idx, si], onset[idx])
    sns.regplot(tmp_stem.loc[idx, si], onset[idx], ax=axes[i],
                label='Pearson \nr={0:.2f} p={1:.3}'.format(r, p))
    axes[i].set_xlabel('{0} {1}'.format('PNT', si))
    axes[i].set_title('{0} (N={1})'.format(disease, tmp_stem.shape[0]))
    axes[i].legend()
fig.tight_layout()
sns.despine()
```

```python
fig.savefig('../data/Analysis/onset_age/20180929/All.PNT_{0}_onset_age.png'.format(si))
fig.savefig('../data/Analysis/onset_age/20180929/All.PNT_{0}_onset_age.pdf'.format(si))
```

```python
brca_stem = stem[stem['disease'] == 'PRAD']
brca_stem = brca_stem[brca_stem.duplicated('patient_id', keep=False)]
```

```python
from scipy.stats import wilcoxon
fig, axes = plt.subplots(1, 2, figsize=(10, 6))
for i in range(2):
    diff = []
    brca_stem = stem[stem['disease'] == disease_list[i]]
    brca_stem = brca_stem[brca_stem.duplicated('patient_id', keep=False)]
    sns.boxplot(data=brca_stem, x='tissue_status', y='mRNAsi', order=['PNT', 'Tumor'], boxprops=dict(alpha=.9), ax=axes[i])
    for patient in brca_stem['patient_id'].unique():
        y1 = brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['tissue_status'] == 'PNT')]['mRNAsi'].values[0]
        y2 = brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['tissue_status'] == 'Tumor')]['mRNAsi'].values[0]
        diff.append(y2 -y1)
        axes[i].plot([0, 1], [y1, y2], ls='--', color='dimgrey')
    s, p = wilcoxon(diff)
    axes[i].text(-0.3, 0.55, 'Wilcoxon p={0:.2E}'.format(p))
    axes[i].set_title(disease_list[i])
    axes[i].set_ylim(0.1, 0.6)
sns.despine()
```

# PNT Tumor Stemness correlation

```python
stem.head()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axes = axes.flatten()
disease_list = ['PRAD', 'BRCA', 'LIHC', 'UCEC', 'THCA', 'LUAD', 'COAD', 'ESCA', 'STAD']
si = 'ENHsi'
for i in range(len(disease_list)):
    pnt_list = []
    tumor_list = []
    brca_stem = stem[stem['cancer.type'] == disease_list[i]]
    brca_stem = brca_stem[brca_stem.duplicated('patient_id', keep=False)]
    for patient in brca_stem['patient_id'].unique():
        try:
            y1 = brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['sample.type'] == 11)][si].values[0]
            y2 = brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['sample.type'] == 1)][si].values[0]
        except IndexError:
            continue
        pnt_list.append(y1)
        tumor_list.append(y2)
    r, p = pearsonr(pnt_list, tumor_list)
    sns.regplot(pnt_list, tumor_list, ax=axes[i], 
                label='Pearson \nr={0:.2f} p={1:.3}'.format(r, p))
    axes[i].legend()
    axes[i].set_title('{0} (N={1})'.format(disease_list[i], len(pnt_list)))
    axes[i].set_xlabel('PNT {0}'.format(si))
    axes[i].set_ylabel('Tumor {0}'.format(si))
fig.tight_layout()
sns.despine()
```

```python
fig.savefig('../data/Analysis/onset_age/20180929/All.PNT_Tumor_{0}_pearsonr.pdf'.format(si))
fig.savefig('../data/Analysis/onset_age/20180929/All.PNT_Tumor_{0}_pearsonr.png'.format(si))
```

```python
for si in ['mRNAsi', 'EREG-mRNAsi']:
    fig, axes = plt.subplots(figsize=(5, 5))
    pnt_list = []
    tumor_list = []
    cancer_type_list = []
    brca_stem = stem
    brca_stem = brca_stem[brca_stem.duplicated('patient_id', keep=False)]
    for patient in brca_stem['patient_id'].unique():
        try:
            y1 = brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['sample.type'] == 11)][si].values[0]
            y2 = brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['sample.type'] == 1)][si].values[0]
            cancer_type_list.append(brca_stem[(brca_stem['patient_id'] == patient) & (brca_stem['sample.type'] == 1)]['cancer.type'].values[0])
        except IndexError:
            continue
        pnt_list.append(y1)
        tumor_list.append(y2)
    r, p = pearsonr(pnt_list, tumor_list)
    sns.scatterplot(pnt_list, tumor_list, ax=axes, hue=cancer_type_list, legend=False)
    sns.regplot(pnt_list, tumor_list, ax=axes, scatter_kws={"alpha": 0},
                label='Pearson \nr={0:.2f} p={1:.3}'.format(r, p))
    axes.legend()
    axes.set_title('{0} (N={1})'.format('PanCancer', len(pnt_list)))
    axes.set_xlabel('PNT {0}'.format(si))
    axes.set_ylabel('Tumor {0}'.format(si))
    #axes.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    fig.tight_layout()
    sns.despine()
    fig.savefig('../data/Analysis/PNT_tumor-correlation/20180929/PanCancer.PNT_Tumor_{0}_pearsonr.pdf'.format(si))
    fig.savefig('../data/Analysis/PNT_tumor-correlation/20180929/PanCancer.PNT_Tumor_{0}_pearsonr.png'.format(si))
```
